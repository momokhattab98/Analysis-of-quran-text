## Reading the files

```python
fopen=open(fpath)
lines=fopen.readlines()
fopen.close()
```

## we need to get the exact data by:
### 1. splitting every row with the (|)
### 2. Deleting the (/n)

```python
lines[0].split("|")
```

```python
new_lines=[]
for line in lines:
  line=line.strip("\n")
  line_split=line.split("|")
  if len(line_split)!=3: continue
  new_lines.append(line_split)

new_lines[:10]

```

```python
sorted_lines=sorted(new_lines,key=lambda x:-len(x[2])) 
```
